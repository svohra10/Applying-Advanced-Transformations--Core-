# Clean Data

```python
# create hero publisher column 
df_info[['Name','Publisher']] = df_info['Hero|Publisher'].str.split('|', expand=True)
```

```python
# drop OG column 
df_info = df_info.drop(columns = ['Hero|Publisher'])
```

```python
first_column = df_info.pop('Name')
```

```python
# Insert Name column as the first column
df_info.insert(0, 'Name', first_column)
```

```python
# Create a sample value
test_measure = df_info.loc[0, 'Measurements']
```

```python
# Display the sample value's datatype and the sample value
print(type(test_measure))
test_measure
```

```python
# fix '' into ""
test_measure = test_measure.replace("'",'"')
test_measure
```

```python
# Convert string to dictionary
test_measure = json.loads(test_measure)
print(type(test_measure))
test_measure
```

```python
# replace all single quotes
df_info['Measurements'] = df_info['Measurements'].str.replace("'",'"')
df_info['Measurements'] = df_info['Measurements'].apply(json.loads)
df_info['Measurements'].head()
```

```python
# Check a sample value after transformation
test_measure = df_info.loc[0, 'Measurements']
print(type(test_measure))
test_measure
```

```python
# Unpack height_weight into a pandas series
height_weight = df_info['Measurements'].apply(pd.Series)
# Display the  series
height_weight.head()
```

```python
# Concat Height and Weight
df_info = pd.concat((df_info, height_weight), axis = 1)
```

```python
df_info.head()
```

```python
# drop transformed column
df_info = df_info.drop(columns=['Measurements'])
```

```python
# Remove spaces and unit of measures
to_replace = [' ', 'cm']
for char in to_replace:
    df_info['Height'] = df_info['Height'].str.replace(char, '', regex=False)
```

```python
# Convert Height column to int datatype
df_info["Height"] =pd.to_numeric(df_info["Height"]).astype(int)
```

```python
df_info["Height"].dtype
```

```python
# Remove spaces and unit of measures
to_replace = [' ', 'kg']
for char in to_replace:
    df_info['Weight'] = df_info['Weight'].str.replace(char, '', regex=False)
```

```python
# Check Weight and Height 
df_info.head()
```

```python
df_info["Weight"] =pd.to_numeric(df_info["Weight"]).astype(int)
df_info["Weight"].dtype
```

```python
# create power series 
new_powers = df_power["Powers"].str.split(",")
```

```python
new_powers
```

```python
# Explode the series into seperate instances
new_powers.explode("Powers")
```

```python
# create and display array of new columns 
cols_to_make = new_powers.explode("Powers").dropna().sort_values().unique()
cols_to_make
```

```python
# Loop through array of new columns to make and create a new column for each
# Indicate true or false depending on if the value was in Powers
for col in cols_to_make:
    df_power[col] = df_power['Powers'].str.contains(col)
```

```python
df_power = df_power.drop(columns=['Powers'])
```
